```python
%load_ext autoreload
%autoreload 2
import numpy as np
import matplotlib.pyplot as plt
from rpca import RobustPCA
from sklearn.decomposition import PCA
```

```python
X.mean(axis=0).shape
```

```python
x_rpca = rpca.fit_transform(X)
x_pca = pca.fit_transform(X)
```
